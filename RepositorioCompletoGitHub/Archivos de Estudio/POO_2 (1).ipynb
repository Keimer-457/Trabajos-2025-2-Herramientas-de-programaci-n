Siempre hay que referenciar el self, para que no de errores, para autoreferenciarse a sí mismo, que sea con los parametros del mismo objeto que se está creando.

```python
class Atraccion:
    def __init__(self, nombre, capacidad, modelo, ticket, musica):
        self.nombre = nombre
        self.capacidad = capacidad
        self.modelo = modelo
        self.ticket = ticket
        self.musica = musica

    # Definimos un método llamado prueba
    def prueba(self):
        print("Prueba de que una función ahora se llama método")
        print(f"Saludo desde la atracción {self.nombre}")

    # Definimos otro método llamado prueba2
    def prueba2(self):
        print("Prueba de otra función que ahora se llama método")
        print(f"Saludo desde la atracción {self.nombre}")

# Creamos tres objetos (instancias) de la clase Atraccion
atraccion1 = Atraccion("Carrusel", "30", "2020", "10000", "Carnaval")
atraccion2 = Atraccion("Carros chocones", "20", "2005", "15000", "Disco")
atraccion3 = Atraccion("Casa del terror", "25", "2020", "12000", "Carnaval")

# Llamamos al método prueba2() sobre el objeto atraccion1
atraccion1.prueba2()

# Llamamos al método prueba() sobre el objeto atraccion2
atraccion2.prueba()
```

Tipos de Herencia en POO:
La herencia es uno de los pilares fundamentales de la POO. Permite que una clase (llamada subclase) herede atributos y métodos de otra clase (llamada superclase). Esto fomenta la reutilización de código y la organización jerárquica.

1. Herencia Simple:
En este tipo de herencia, una clase hereda de una sola superclase. Por ejemplo:

```python
class Juego(atraccion):  # Juego hereda de Atraccion
    def __init__(self, nombre, capacidad, modelo, ticket, musica, dificultad):
        super().__init__(nombre, capacidad, modelo, ticket, musica)
        self.dificultad = dificultad

    def mostrar_dificultad(self):
        print(f"La dificultad del juego {self.nombre} es {self.dificultad}")
```

Aquí, la clase Juego hereda de Atraccion.
Usamos super() para llamar al constructor de la clase padre (Atraccion) y añadir nuevos atributos (dificultad).

2. Herencia Múltiple:
Una clase puede heredar de más de una superclase. Esto permite combinar funcionalidades de varias clases. Por ejemplo:

```python
class Tecnologia:
    def __init__(self, version):
        self.version = version

    def mostrar_version(self):
        print(f"Versión tecnológica: {self.version}")

class JuegoInteractivo(Atraccion, Tecnologia):  # Herencia múltiple
    def __init__(self, nombre, capacidad, modelo, ticket, musica, dificultad, version):
        Atraccion.__init__(self, nombre, capacidad, modelo, ticket, musica)
        Tecnologia.__init__(self, version)
        self.dificultad = dificultad

    def mostrar_dificultad(self):
        print(f"La dificultad del juego interactivo {self.nombre} es {self.dificultad}")


```

JuegoInteractivo hereda de Atraccion y Tecnologia.
Es necesario llamar explícitamente a los constructores de ambas superclases.


3. Herencia Multinivel:
En este caso, una clase hereda de otra clase que ya es una subclase. Por ejemplo:

```python
class AtraccionEspecial(Atraccion):  # Primera herencia
    def __init__(self, nombre, capacidad, modelo, ticket, musica, efectos_especiales):
        super().__init__(nombre, capacidad, modelo, ticket, musica)
        self.efectos_especiales = efectos_especiales

class AtraccionVIP(AtraccionEspecial):  # Segunda herencia
    def __init__(self, nombre, capacidad, modelo, ticket, musica, efectos_especiales, vip_access):
        super().__init__(nombre, capacidad, modelo, ticket, musica, efectos_especiales)
        self.vip_access = vip_access
```

AtraccionVIP hereda de AtraccionEspecial, que a su vez hereda de Atraccion.


4. Herencia Jerárquica:
Cuando varias clases heredan de una misma superclase. Por ejemplo:

```python
class Carrusel(Atraccion):
    pass

class CarrosChocones(Atraccion):
    pass

class CasaDelTerror(Atraccion):
    pass
```

Carrusel, CarrosChocones y CasaDelTerror heredan de la misma clase Atraccion.


5. Herencia Híbrida:
Es una combinación de dos o más tipos de herencia. Por ejemplo:

```python
class JuegoVirtual(Atraccion, Tecnologia):
    def __init__(self, nombre, capacidad, modelo, ticket, musica, version, realidad_virtual):
        Atraccion.__init__(self, nombre, capacidad, modelo, ticket, musica)
        Tecnologia.__init__(self, version)
        self.realidad_virtual = realidad_virtual
```

Aquí, JuegoVirtual utiliza tanto herencia múltiple como multinivel.


```python
class Atraccion:
    def __init__(self, nombre, capacidad, modelo, ticket, musica):
        # Estos son los atributos de la clase Atraccion
        self.nombre = nombre
        self.capacidad = capacidad
        self.modelo = modelo
        self.ticket = ticket
        self.musica = musica

    # Método 1
    def prueba(self):
        print("Prueba de que una función ahora se llama método")
        print(f"Saludo desde la atracción {self.nombre}")

    # Método 2
    def prueba2(self):
        print("Prueba de otra función que ahora se llama método")
        print(f"Saludo desde la atracción {self.nombre}")


# Clase hija: hereda de Atraccion
class Juego(Atraccion):
    def __init__(self, nombre, capacidad, modelo, ticket, musica, dificultad):
        # Llamamos al constructor de la clase padre con super()
        super().__init__(nombre, capacidad, modelo, ticket, musica)

        # Agregamos un nuevo atributo propio de la clase hija
        self.dificultad = dificultad

    # Método nuevo de la clase hija
    def mostrar_dificultad(self):
        print(f"La dificultad del juego {self.nombre} es {self.dificultad}")

# Creamos un objeto de la clase hija (Juego) y
# vemos cómo puede usar tanto los métodos heredados como los nuevos.

# Creamos un objeto de la clase hija
juego1 = Juego("Carros chocones", "20", "2005", "15000", "Disco", "Fácil")

# Llamamos a los métodos heredados desde la clase padre
juego1.prueba()
juego1.prueba2()

# Llamamos al método nuevo de la clase hija
juego1.mostrar_dificultad()

```

El objeto juego1 ahora tiene seis atributos en total:

5 heredados

1 nuevo

##Encapsulamiento

El encapsulamiento significa proteger los datos de un objeto, de modo que no se puedan modificar directamente desde fuera de la clase.
Esto se hace para controlar el acceso y evitar errores o usos indebidos.

En Python, no existe una protección total como en otros lenguajes (por ejemplo, private o protected),
pero sí se sigue una convención usando guiones bajos (_ y __) para indicar el nivel de acceso.

```python
#sin encapsular:
class Atraccion:
    def __init__(self, nombre, capacidad, ticket):
        self.nombre = nombre
        self.capacidad = capacidad
        self.ticket = ticket
# Crea una atracción
atraccion1 = Atraccion("Montaña Rusa", 30, 15000)

# Accedemos directamente a los atributos
print(atraccion1.capacidad)   # Se puede leer
atraccion1.capacidad = 100    # También se puede modificar sin control
print(atraccion1.capacidad)
```

```python
# encapsulado
class Atraccion:
    def __init__(self, nombre, capacidad, ticket):
        self._nombre = nombre       # Atributo protegido
        self._capacidad = capacidad # Atributo protegido
        self._ticket = ticket       # Atributo protegido

    def mostrar_info(self):
        print(f"Atracción: {self._nombre}, Capacidad: {self._capacidad}, Ticket: ${self._ticket}")

atraccion2 = Atraccion("Carrusel", 20, 10000)
atraccion2.mostrar_info()


atraccion2._capacidad = 200  # Técnicamente se puede, pero no se recomienda
```

Encapsulamiento con atributos privados

Si realmente queremos proteger el atributo, usamos doble guion bajo (__).
Esto activa el name mangling: Python cambia el nombre internamente para que no se pueda acceder fácilmente.

```python
class Atraccion:
    def __init__(self, nombre, capacidad, ticket):
        self.__nombre = nombre        # Atributo privado
        self.__capacidad = capacidad  # Atributo privado
        self.__ticket = ticket        # Atributo privado

    def mostrar_info(self):
        print(f"Atracción: {self.__nombre}, Capacidad: {self.__capacidad}, Ticket: ${self.__ticket}")
```

```python
atraccion3 = Atraccion("Casa del Terror", 15, 12000)
atraccion3.mostrar_info()
```

Uso de getters y setters

Para acceder o modificar los atributos privados de manera controlada, se crean métodos llamados getters (leer) y setters (modificar).

```python
class Atraccion:
    def __init__(self, nombre, capacidad, ticket):
        self.__nombre = nombre
        self.__capacidad = capacidad
        self.__ticket = ticket

    # Getter: permite leer el valor de capacidad
    def get_capacidad(self):
        return self.__capacidad

    # Setter: permite cambiar la capacidad, pero con control
    def set_capacidad(self, nueva_capacidad):
        if nueva_capacidad > 0:
            self.__capacidad = nueva_capacidad
        else:
            print("La capacidad debe ser mayor que cero")

    # Método para mostrar información
    def mostrar_info(self):
        print(f"Atracción: {self.__nombre}, Capacidad: {self.__capacidad}, Ticket: ${self.__ticket}")
```

```python
atraccion4 = Atraccion("Tobogán", 10, 8000)

# Mostrar información
atraccion4.mostrar_info()

# Intentamos modificar la capacidad con el setter
atraccion4.set_capacidad(25)
atraccion4.mostrar_info()

# Intentamos poner un valor no válido
atraccion4.set_capacidad(-5)
```

El encapsulamiento protege los datos del objeto.

Se controlan los cambios con métodos especiales (get/set).

Mejora la seguridad y consistencia del código.
